```python
def tcs(i,s):
    for i in range(i,s+1):
        yield i

bbc = tcs(11,21)
print(bbc)
print(next(bbc))
print(next(bbc))
print(next(bbc))
print(next(bbc))
print(next(bbc))
print(next(bbc))
print(next(bbc))
print(next(bbc))
print(next(bbc))
print(next(bbc))
```

```python
def counter(f):
    print(f)
    if counter == 0:
        return
    else:
        counter(f+1)
counter(10)
```

```python
def countdown(n):
    print(n)
    if n == 0:
        return             # Terminate recursio
    else:
        countdown(n - 1) 
countdown(5)
```
